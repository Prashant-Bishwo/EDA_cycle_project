```python
# Select all categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns 
# Encode categorical columns 
df[categorical_cols] = df[categorical_cols].apply(lambda col: col.astype('category').cat.codes) 
# Verify encoding
print("\nEncoded Data:\n", df) 
corr_matrix = df.corr()
# Calculate correlation matrix corr_matrix = df.corr()
 # Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
# show graphs
plt.show()

# encode non numerical data into numerical data form . 
```

```python
# Using One-Hot Encoding for categorical columns to change to numeric
df_encoded = pd.get_dummies(df, drop_first=True)

# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()

# Setting the style
sns.set(style="white")

# Creating the heatmap
plt.figure(figsize=(40, 40))  # Adjusting the figure size
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm',
                       square=True, cbar_kws={"shrink": .8}, linewidths=.5,
                       annot_kws={"size": 10, "weight": "bold"},  # Customizing annotation size and weight
                       cbar=True)  # Show color bar

# Customizing the plot
heatmap.set_title("Correlation Heatmap", fontsize=22, weight='bold')
plt.xticks(fontsize=14, rotation=45)  # Rotating x-axis labels for better readability
plt.yticks(fontsize=14)
plt.tight_layout()

# Displaying the plot
plt.show()
```

The heatmap shows the Pearson correlation coefficients between all pairs of variables in the dataset.The value ranges from -1 to 1, where 1 implies a perfect positive correlation, -1 implies a perfect negative correlation, and 0 implies no correlation.

The correlation between age and G3 is -0.16, which shows that a negative correlation between age and G3 would imply that older students are slightly more likely to have lower final grades compared to younger students.

## Objective
The primary objective of this analysis is to examine the relationship between age and G3 among students.

### Gender Analysis

### Box Plot Insights

The analysis reveals a weak negative correlation (-0.16) between age and G3, Older students are  more likely to have lower final grades compared to younger, though this is not a strong predictor.
